##### NOTE
How this works is that each animal will be assigned an option based on how I believe they would answer. So for example, if the question was "You come across someone bullying a friend. What would you do?" and my answer choices are:

- (A) Ignore and keep on walking
- (B) Fight the bully
- (C) Join the bully
- (D) Report the situation to a teacher and walk away

Then based on the animal's given characteristics (once again, subjective), then they will be given points for that answer choice. So if the user chose "(B) fight the bully", then Sparklemuffin and Blind Cavefish would get a point added. By the end of the game, the animal with the most points will be the winner. In the case of a tie...I'll work that out later.

##### Defining the animal's characteristics (this is subjective and how I believe what personality each animal will have!)

- Naked molerat ~ Hates life
